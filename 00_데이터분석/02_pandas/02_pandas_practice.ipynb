```python
import pandas as pd

stock_info = pd.read_csv("../00_data/stock_daily_prices.csv", index_col="Date")
stock_info
```

# 문제1. 각 종목의 최고가와 최저가 찾기

```python
max_prices = stock_info.max()
min_prices = stock_info.min()

print(f"종목별 최고가\n{max_prices}\n\n종목별 최저가\n{min_prices}")
```

# 문제2. s&p500과 테슬라 간의 상관관계 분석

```python
sp500_tsla = stock_info[["sp500", "TSLA"]]
sp500_tsla_corr = sp500_tsla.corr()

print(f"s&p500과 테슬라의 상관관계: {sp500_tsla_corr.iloc[0]["TSLA"]:.3f}")
```

# 문제3. 특정 기간(2012-01-12 ~ 2012-01-25) 동안 가장 많이 상승한 종목 찾기

```python
change_pct = (stock_info.loc["2012-01-25"] - stock_info.loc["2012-01-12"]) / stock_info.loc["2012-01-12"] * 100

print(f"2012-01-12 ~ 2012-01-25 동안 가장 많이 상승한 종목: {change_pct.idxmax()}, 상승률: {change_pct.max():.3f}%")
```
